```python
import requests
from bs4  import BeautifulSoup

url = "https://comic.naver.com/webtoon/weekday"

res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')

cartoons = soup.find('ol', {"id": "realTimeRankFavorite"})
ranks = cartoons.select('li > a')
for rank in ranks:
    print(rank.get_text())
```

```python
prev_rank = soup.find_all("ol", {"id": "realTimeRankFavorite"}) #find all 은 여러개를 받아오므로 사용하고 싶으면 [0]번째 인덱스만 가져오기.
ranks = prev_rank[0].find_all("li")
for r in ranks:
    print(r.find("a")["title"])
```

# 제목, url, 평점 나타내기

```python
import requests
from bs4  import BeautifulSoup

url = "https://comic.naver.com/webtoon/list?titleId=675554"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')

titles= soup.find_all('td', {'class' : 'title'})

urls = soup.find_all('td', {'class' : 'title'})

stars = soup.find_all('div', {'class': 'rating_type'})
for title, url, star in zip(titles, urls, stars):
    st = star.find('strong')
    url = url.find('a')
    url = url['href']
    print("제목 :{}, url: https://comic.naver.com{}, 평점 : {}".format(title.get_text().strip(), url, st.get_text()))
```

# 평점만 건들기

```python
import requests
from bs4  import BeautifulSoup

url = "https://comic.naver.com/webtoon/list?titleId=675554"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')

stars = soup.find_all('div', {'class': 'rating_type'})
num = 0
for star in stars:
    print(star.find('strong').get_text())
    num += float(star.find('strong').get_text())
num_rate = num / len(stars)
print(f'총합 : {num}, 평균: {num_rate}')
```

```python
import requests
from bs4  import BeautifulSoup

url = "https://www.coupang.com/np/search?component=&q=%EB%85%B8%ED%8A%B8%EB%B6%81&channel=user"
#headers를 쓰면 조건같은 느낌을 부여했기에 먹통이 좀 괜찮아짐.
headers = {'user-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
res = requests.get(url, headers = headers)

soup = BeautifulSoup(res.text, 'lxml')
```

```python
titles = soup.find_all('div', {'class' : 'name'})
prices = soup.find_all('strong', {'class' : 'price-value'})
stars = soup.find_all('em', {'class': 'rating'})
reviews = soup.find_all('span', {'class': 'rating-total-count'})
for title, price, star, review in zip(titles, prices, stars, reviews):
    print(title.text, price.text,'원, 평점: {}, 리뷰수 {}\n'.format(star.text, review.text))
```

```python
import requests
from bs4  import BeautifulSoup

for i in range(1, 6):
    url = f"https://www.coupang.com/np/search?q=%EB%85%B8%ED%8A%B8%EB%B6%81&channel=user&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=36&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page={i}&rocketAll=false&searchIndexingToken=1=4&backgroundColor="
    headers = {'user-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    res = requests.get(url, headers = headers)

    soup = BeautifulSoup(res.text, 'lxml')
    products = soup.find_all('li' , {'class' : 'search-product'})
    for product in products:
        title = product.find('div', {'class' : 'name'})
        if '삼성' not in title.text:
            continue
        price = product.find('strong', {'class' : 'price-value'})
        star = product.find('em', {'class': 'rating'})
        review = product.find('span', {'class': 'rating-total-count'})
        ur = product.find('a')
        ur = ur['href']
        ad = product.find('span', {'class' : 'ad-badge-text'})
        if ad:
            print('\n!!!광고상품제외!!!\n')
            continue
        if (star and review and int(review.text[1:-1]) >= 100 and float(star.text) >= 4.5):
            print(f'페이지 : ({i})\n {title.text} \n가격: {price.text}, 평점: {star.text}, 리뷰수: {review.text}\n 링크 : https://www.coupang.com{ur}\n', '*' * 50)
```

```python
import requests
from bs4  import BeautifulSoup

url = "https://search.daum.net/search?w=tot&q=2020%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
headers = {'user-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
res = requests.get(url, headers = headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')
print(soup)
imgs = soup.find_all('img', {'class' : 'thumb_img'})
for i, img in enumerate(imgs):
    img_url = img['src']
    img_res = requests.get(img_url)
    img_res.raise_for_status
    with open ('movie_{}'.format(i), 'wb') as file:
        file.write(img.content)
```
